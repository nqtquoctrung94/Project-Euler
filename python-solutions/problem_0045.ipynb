## Triangular, Pentagonal, and Hexagonal

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

| **Polygonal** | **Function**      | **Values**                |
|---------------|-------------------|---------------------------|
|      Triangle | $T_n = n(n+1)/2$  | $1, 3, 6, 10, 15, \dots$  |
|    Pentagonal | $P_n = n(3n-1)/2$ | $1, 5, 12, 22, 35, \dots$ |
|     Hexagonal | $H_n = n(2n-1)$   | $1, 6, 15, 28, 45, \dots$ |

It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.


# Solution

Every hexagonal number is a triangular number. 

Proof: https://en.wikipedia.org/wiki/Polygonal_number#Every_hexagonal_number_is_also_a_triangular_number

```python
from snippet.euler_lib import is_pentagonal_number
```

```python
def s_gonal_number(side: int, index: int) -> int:
    """Ref: https://en.wikipedia.org/wiki/Polygonal_number"""
    return (side - 2)*(index*(index-1)//2) + index
```

```python
def solution():
    hex_index = 144
    while True:
        hex_number = s_gonal_number(side=6, index=hex_index)
        if is_pentagonal_number(hex_number):
            return hex_number
        hex_index += 1
```
