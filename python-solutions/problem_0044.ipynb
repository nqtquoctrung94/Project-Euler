# [Question 44](https://projecteuler.net/problem=44)

## Pentagon Numbers
Pentagonal numbers are generated by the formula, $P_n=n(3n-1)/2$. The first ten pentagonal numbers are:<br>
$$1, 5, 12, 22, 35, 51, 70, 92, 117, 145, \dots$$

It can be seen that $P_4 + P_7 = 22 + 70 = 92 = P_8$. However, their difference, $70 - 22 = 48$, is not pentagonal.<br>
Find the pair of pentagonal numbers, $P_j$ and $P_k$, for which their sum and difference are pentagonal and $D = |P_k - P_j|$ is minimised; what is the value of $D$?<br>


# Solution

```python
from euler_lib import is_pentagonal_number
```

```python
def is_pentagonal_pair(num1: int, 
                       num2: int) -> bool:
    return is_pentagonal_number(num1+num2) and is_pentagonal_number(abs(num1-num2))
```

```python
def min_different_number_vs_list(input_num: int, 
                                 input_list: list, 
                                 default_highest: int) -> int:
    return min(
        [abs(input_num - num) for num in input_list if is_pentagonal_pair(input_num, num)], 
        default= default_highest
    )
```

```python
def pentagonal_number(index: int) -> int:
    return index*(3*index-1)//2
```

```python
def create_pentagonal_list(limit: int) -> list:
    p_list = []
    n = 1
    while pentagonal_number(n) < limit:
        p_list.append(pentagonal_number(n))
        n+=1
    return p_list
```

```python
def solution():
	limit = 10000000
	pentagonal_list = create_pentagonal_list(limit)
	return min(
		[min_different_number_vs_list(value, pentagonal_list[:index], limit) for index,value in enumerate(pentagonal_list)],
		default= limit
	)
```
